```python
# Token Loss Visualization Example
import os
import sys
from tqdm import tqdm
# Enable automatic module reloading (great for development)
%load_ext autoreload
%autoreload 2

# Add the directory with your visualization module to the path
sys.path.append("/Users/liam/quests/lsoc-psych/datasets")
from dm_math_plots import load_token_data, load_losses_data, create_loss_visualizations, extract_token_losses, process_all_data

```

```python
# Path to your data files
model_sizes = ['14m', '31m', '70m', '160m', '410m', '1b', '1.4b', '2.8b', '6.9b']
token_file_path = "/Users/liam/quests/lsoc-psych/datasets/experiments/EXP000/trajectories/tokens/dm_mathematics_tokens.pkl"
data_dict = {}
for model_size in tqdm(model_sizes):
    few_shot_file_path = f"/Users/liam/quests/lsoc-psych/datasets/experiments/EXP000/trajectories/csv/{model_size}_dm_mathematics_few_shot.csv"
    zero_shot_file_path = f"/Users/liam/quests/lsoc-psych/datasets/experiments/EXP000/trajectories/csv/{model_size}_dm_mathematics_zero_shot.csv"
    token_data = load_token_data(token_file_path)
    few_shot_losses = load_losses_data(few_shot_file_path)
    zero_shot_losses = load_losses_data(zero_shot_file_path)

    all_data = process_all_data(token_data, few_shot_losses, zero_shot_losses)
    data_dict[model_size] = all_data

    
# few_shot_file_path = "/Users/liam/quests/lsoc-psych/datasets/experiments/EXP000/trajectories/csv/1.4b_dm_mathematics_few_shot.csv"
# zero_shot_file_path = "/Users/liam/quests/lsoc-psych/datasets/experiments/EXP000/trajectories/csv/1.4b_dm_mathematics_zero_shot.csv"

```

```python
plotting_path = "/Users/liam/quests/lsoc-psych/datasets/plotting/dm_math_category_losses"
tol=0.3
for model_size, data in data_dict.items():
    fig = create_loss_visualizations(
        data['few_shot_full'],
        data['few_shot_answer'],
        data['zero_shot_full'],
        data['zero_shot_answer'],
        data['full_merged'],
        data['answer_merged'],
        use_log_scale=True,
        model_size=model_size,
        axes_ranges = {(1,1): [0-tol, 13],
                       (1,2): [0-tol, 13],
                       (1,3): [-3,2],
                       (2,1): [0-tol, 13],
                       (2,2): [0-tol, 13],
                       (2,3): [-6,1]}
    )

    plot_path = f"{plotting_path}/{model_size}_category_losses.png"
    fig.write_image(plot_path, scale=2)
    fig.write_html(f"{plotting_path}/{model_size}_category_losses.html")
    
    # Display or save the figure
#fig.show()
```
